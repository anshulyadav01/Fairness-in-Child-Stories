```python
df.head()
```

```python
df1 = df.loc[df['Author Gender'] == 'Male']
```

```python
count_event_m = {}
count_event_f = {}
t_count_event_m = 0
t_count_event_f = 0
for index, row in df1.iterrows():
    if row['Gender'] ==  'M':
      t_count_event_m += 1
      if (row['Gender'], row['Event']) not in count_event_m:
        count_event_m[(row['Gender'], row['Event'])] = 1
      else:
        count_event_m[(row['Gender'], row['Event'])] += 1
    else:
      t_count_event_f += 1
      if (row['Gender'], row['Event']) not in count_event_f:
        count_event_f[(row['Gender'], row['Event'])] = 1
      else:
        count_event_f[(row['Gender'], row['Event'])] += 1
male_odds = sorted(count_event_m.items() ,  key=lambda x: x[1],reverse= 1)
female_odds = sorted(count_event_f.items() ,  key=lambda x: x[1],reverse= 1)


male_odd_ratio = {}
for (a,b) in count_event_m:
  if count_event_m[(a,b)]>= 2:
    if ('F', b) in count_event_f and count_event_f[('F', b)] >=2:
      male_odd_ratio[b] = count_event_m[(a,b)]/ t_count_event_m / count_event_f[('F', b)] * t_count_event_f
    elif count_event_m[(a,b)] > 20:
      male_odd_ratio[b] = 100
male_odds_ratio = sorted(male_odd_ratio.items() ,  key=lambda x: x[1],reverse= 1)
female_odd_ratio = {}
for (a,b) in count_event_f:
  if count_event_f[(a,b)]>= 2:
    if ('M', b) in count_event_m and count_event_m[('M', b)] >=2:
      female_odd_ratio[b] = count_event_f[(a,b)]/ t_count_event_f / count_event_m[('M', b)] * t_count_event_m
    elif count_event_f[(a,b)] > 20:
      female_odd_ratio[b] = 100
female_odds_ratio = sorted(female_odd_ratio.items() ,  key=lambda x: x[1],reverse= 1)
male_odds_list = []
for ((a,b), c) in male_odds:
  male_odds_list.append([b, c/t_count_event_m])

female_odds_list = []
for ((a,b), c) in female_odds:
  female_odds_list.append([b, c/t_count_event_f])


```

```python
import csv 

with open('odd_ratio_female_M_author.csv', 'w') as f: 
    write = csv.writer(f) 
    write.writerows(female_odds_ratio) 
with open('odd_ratio_male_M_author.csv', 'w') as f: 
    write = csv.writer(f) 
    write.writerows(male_odds_ratio) 
with open('odds_male_M_author.csv', 'w') as f: 
    write = csv.writer(f) 
    write.writerows(male_odds_list) 
with open('odds_female_M_author.csv', 'w') as f: 
    write = csv.writer(f) 
    write.writerows(female_odds_list) 
```
