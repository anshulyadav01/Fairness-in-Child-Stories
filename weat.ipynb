```python
from wefe import WordEmbeddingModel, Query, WEAT
import gensim.downloader as api
import nltk
from nltk.stem import WordNetLemmatizer
wnl = WordNetLemmatizer()
nltk.download('wordnet')
word2vec_model = WordEmbeddingModel(api.load('word2vec-google-news-300'))
```

```python
import pandas as pd
dct = {}
male_e = {}
female_e = {}
```

```python
def calculate_weat(male, female):
  female_list=['lady','female','girl','wife','Se√±ora','gal','sister','sheila','broad','maiden','damsel','gentlewoman','daughter', 'hers', 'grandma', 'grandmother', 'aunt', 'sis', 'niece', 'mother', 'she', 'her', 'granny', 'granddaughter', 'girlfriend', 'woman', 'mom','miss','mrs.','ms']
  male_list=['man','gentleman','boy','youth','guy','fellow','gent','bloke','chap','hombre','dude', 'godfather', 'grandson', 'stepbrother', 'sir', 'he', 'uncle', 'male', 'soninlaw', 'boyfriend', 'brother', 'grandpa', 'him', 'nephew', 'son', 'papa', 'exboyfriend', 'granddad', 'husband', 'stepson', 'dad', 'fatherinlaw', 'daddy', 'stepdad', 'father', 'grandfather', 'bro', 'his','Mr.']
  gender_occupation_query = Query([male_list, female_list],
                                  [male, female],
                                  ['Male Pronouns', 'Female pronouns'],
                                  ['Male Events', 'Female Events'])
  weat = WEAT()
  results = weat.run_query(gender_occupation_query, word2vec_model)
  return results['weat']

m = pd.read_csv("odd_ratio_male.csv", header=None)
f = pd.read_csv("odd_ratio_female.csv", header=None)
male = m[0].values.tolist()[:10]
female = f[0].values.tolist()[:10]

male_e['F_author'] = male
female_e['F_author'] = female
dct['General'] = calculate_weat(male, female)
print(male_e)
print(female_e)
print(dct)
```

```python
# Outputs the WEAT Scores by Genre Category for top 10 Events for each Gender
weat = pd.DataFrame.from_dict([male_e, female_e, dct]).T.
weat.rename(columns={0: "Top 10 Male Events", 1:"Top 10 Female Events", 2: "WEAT Score"})
```
